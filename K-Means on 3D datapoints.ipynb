```python
#--------------------------------------MACHINE LEARNING | ASSIGNMENT 3----------------------------------------------#

#------------------------------------------------CLUSTERING---------------------------------------------------------#

#Creating 3D data d_1,d_2,d_3

import pandas as pd
import numpy as np
from matplotlib import pyplot
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
from copy import deepcopy

fig = pyplot.figure()
ax = Axes3D(fig)

# Euclidean Distance Caculator
def dist(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)

for c, m in [('r', 'o'), ('b', '^')]:
    d_1=np.random.rand(1,500)
    d_2=np.random.rand(1,500)
    d_3=np.random.rand(1,500)
    ax.scatter(d_1.T, d_2.T, d_3.T, c=c, marker=m)
    
ax.set_xlabel('X Vector')
ax.set_ylabel('Y Vector')
ax.set_zlabel('Z Vecttor')

plt.show()

vector=np.array(list(zip(d_1,d_2,d_3)))
print(vector.shape)
```

```python
f_1=d_1.T
f_2=d_2.T
f_3=d_3.T
```

```python
Xl=list(zip(f_1,f_2,f_3))
```

```python
np.asarray(Xl).reshape((500,3))
```

```python
Xl=np.array(Xl)
X=Xl[:,:,0]
```

```python
#-------------------------------------------------K=3-------------------------------------------------#

C1=np.random.rand(3,3)
C_old = np.zeros(C1.shape)
clusters = np.zeros(len(X))
error = dist(C1, C_old, None)

# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i], C1)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C1)
    # Finding the new centroids by taking the average value
    for i in range(3):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C1[i] = np.mean(points, axis=0)
    error = dist(C1, C_old, None)
```
